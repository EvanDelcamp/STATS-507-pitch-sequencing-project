```python
# IF ANY PLAYER FAILS, JUST RESTART KERNEL AND TRY AGAIN - SOMETIMES THE SERVER HAS MOMENTARY ISSUES
from pybaseball import statcast_pitcher
import pandas as pd
import numpy as np

# ---- CONFIG ----
start_date = '2023-03-01'
end_date = '2024-09-30'  # Adjust as needed

# ---- STEP 1: Pull Statcast Data ----
all_pitcher_data = []

for name, player_id in pitcher_ids.items():
    print(f"Pulling data for {name} (ID: {player_id})...")
    try:
        df = statcast_pitcher(start_date, end_date, player_id=player_id)
        df['pitcher_name'] = name  
        all_pitcher_data.append(df)
    except Exception as e:
        print(f"Failed for {name}: {e}")

# Combine all into a single DataFrame
df = pd.concat(all_pitcher_data, ignore_index=True)

print(f"\nCombined data shape: {df.shape}")

# Filter out Spring Training Data
df = df[df['game_type'] == 'R']

# ---- STEP 2: Sort for sequence alignment ----
df = df.sort_values(['game_pk', 'at_bat_number', 'pitch_number'])

# ---- STEP 3: Calculate pitch_count_in_game (fatigue) ----
df['pitch_count_in_game'] = df.groupby(['game_pk', 'pitcher']).cumcount() + 1

# Fielder 2 is catcher
df['catcher_id'] = df['fielder_2']

# ---- STEP 4: Runner-on-base binary flags ----
df['runner_on_1b'] = df['on_1b'].notnull().astype(int)
df['runner_on_2b'] = df['on_2b'].notnull().astype(int)
df['runner_on_3b'] = df['on_3b'].notnull().astype(int)

# ---- STEP 5: Handedness binary flags ----
df['pitcher_throws_left'] = (df['p_throws'] == 'L').astype(int)
df['batter_hits_left'] = (df['stand'] == 'L').astype(int)

# ---- STEP 6: Strike indicator (1 = strike, 0 = ball) ----
# Define half the plate width (in feet)
half_plate_width = 17 / 12 / 2  # â 0.708

# 1 = in strike zone, 0 = outside
df['is_strike'] = (
    (df['plate_x'].abs() <= half_plate_width) &
    (df['plate_z'] >= df['sz_bot']) &
    (df['plate_z'] <= df['sz_top'])
).astype(int)

# This original code was not what was intended. We arent concerned with batter outcomes on the pitch - we want to know if the pitcher threw a 
# pitch in the zone or out of the zone.
#strike_events = [
#    'called_strike', 'swinging_strike', 'foul', 'foul_tip',
#    'swinging_strike_blocked'
#]
#df['is_strike'] = df['description'].isin(strike_events).astype(int)

# ---- STEP 7: Determine pitching team ----
df['pitching_team'] = np.where(
    df['inning_topbot'] == 'Top',
    df['home_team'],
    df['away_team']
)

# ---- STEP 8: Score differential from pitcher's team perspective ----
df['score_differential'] = df['post_home_score'] - df['post_away_score']
df['score_differential'] = np.where(
    df['pitching_team'] == df['home_team'],
    df['score_differential'],
    -df['score_differential']
)

# ---- STEP 9: Select relevant columns ----
feature_df = df[[
    'game_pk', 'at_bat_number',
    'pitcher', 'catcher_id', 'events', 'description',
    'pitch_type', 'zone', 'is_strike',
    'batter_hits_left', 'pitcher_throws_left',
    'balls', 'strikes', 'outs_when_up',
    'pitch_number', 'pitch_count_in_game',
    'runner_on_1b', 'runner_on_2b', 'runner_on_3b',
    'score_differential','plate_x', 'plate_z', 'sz_top', 'sz_bot'
]].copy()

# ---- STEP 10: Rename columns for consistency ----
feature_df.rename(columns={
    'balls': 'ball_count',
    'strikes': 'strike_count',
    'outs_when_up': 'outs',
    'pitch_number': 'pitch_number_in_atbat',
    'pitch_type': 'prev_pitch_type',
    'zone': 'prev_pitch_zone'
}, inplace=True)

# ---- DONE ----
# print(feature_df.head())
display(feature_df)
```

```python
# Step 1: Count total pitches per pitcher
pitcher_totals = df.groupby('pitcher')['pitch_type'].count().rename('total_pitches')

# Step 2: Count pitch type occurrences per pitcher
pitch_type_counts = df.groupby(['pitcher', 'pitch_type']).size().rename('count').reset_index()

# Step 3: Merge and calculate usage rate
pitch_usage = pitch_type_counts.merge(pitcher_totals, on='pitcher')
pitch_usage['usage_rate'] = pitch_usage['count'] / pitch_usage['total_pitches']

# Step 4: Filter and create arsenal dict. Only pitches used more than 2.5 % are considered viable for the model to predict.
pitch_usage_filtered = pitch_usage[pitch_usage['usage_rate'] > 0.025]
arsenal_dict = (
    pitch_usage_filtered
    .groupby('pitcher')['pitch_type']
    .apply(list)
    .to_dict()
)

display(arsenal_dict)
```

```python
def assign_custom_zone(row):
    z_center = (row['sz_top'] + row['sz_bot']) / 2
    z = row['plate_z']
    x = row['plate_x']
    
    if pd.isna(x) or pd.isna(z) or pd.isna(z_center):
        return np.nan  # handle missing data
    
    if abs(x) < 0.5 and abs(z - z_center) < 0.5:
        return 'middle'
    elif z >= z_center:
        return 'top_left' if x <= 0 else 'top_right'
    else:
        return 'bottom_left' if x <= 0 else 'bottom_right'

feature_df['custom_zone'] = feature_df.apply(assign_custom_zone, axis=1)

zone_to_class = {
    'middle': 0,
    'top_left': 1,
    'top_right': 2,
    'bottom_left': 3,
    'bottom_right': 4
}

feature_df['custom_zone_class'] = feature_df['custom_zone'].map(zone_to_class)

display(feature_df)
```

```python
# Build Class for Data Loader
import torch
from torch.utils.data import Dataset

class PitchSequenceDataset(Dataset):
    def __init__(self, sequences, y_pitch_type, y_zone, y_is_strike, input_feature_cols):
        self.sequences = sequences
        self.y_pitch_type = y_pitch_type
        self.y_zone = y_zone
        self.y_is_strike = y_is_strike
        self.input_feature_cols = input_feature_cols

    def __len__(self):
        return len(self.sequences)

    # Extact items
    def __getitem__(self, idx):
        seq_df = self.sequences[idx]
        seq = seq_df[self.input_feature_cols].values.astype(np.float32)
        target_pitch_type = self.y_pitch_type[idx]
        target_zone = self.y_zone[idx]
        target_is_strike = self.y_is_strike[idx]

        # Extract pitcher ID index (assumes constant within sequence)
        pitcher_id = torch.tensor(seq_df['pitcher_id_idx'].iloc[0], dtype=torch.long)

        return (
            torch.tensor(seq),
            torch.tensor(target_pitch_type),
            torch.tensor(target_zone),
            torch.tensor(target_is_strike),
            pitcher_id
        )
```

```python

from torch.nn.utils.rnn import pad_sequence
import torch

def collate_fn(batch):
    # Custom collate function for DataLoader to handle variable-length sequences
    # Each batch item is a tuple: (input_seq, pitch_type, zone, strike, pitcher_id_idx)

    # batch: list of (input_seq, pitch_type, zone, strike, pitcher_id_idx)
    # Sort batch by sequence length in descending order for packing sequences before feeding them into an LSTM
    batch.sort(key=lambda x: len(x[0]), reverse=True)  # Sort by sequence length

    # Separate out individual components from each batch item
    input_seqs = [item[0] for item in batch]
    lengths = torch.tensor([len(seq) for seq in input_seqs], dtype=torch.long)
    padded_seqs = pad_sequence(input_seqs, batch_first=True)  # (batch_size, max_len, input_dim)

    # Pad sequences
    pitch_targets = torch.tensor([item[1] for item in batch])
    zone_targets = torch.tensor([item[2] for item in batch])
    strike_targets = torch.tensor([item[3] for item in batch], dtype=torch.float32)
    pitcher_ids = torch.tensor([item[4] for item in batch], dtype=torch.long)

    return padded_seqs, lengths, pitch_targets, zone_targets, strike_targets, pitcher_ids
```

```python
# Import training data into data loader
from torch.utils.data import DataLoader

# Since pitcher ID is embedded it doesn't need to be a feature column.
input_feature_cols = [
    'prev_pitch_type', 'ball_count', 'strike_count', 'outs',
    'pitch_number_in_atbat', 'pitch_count_in_game',
    'runner_on_1b', 'runner_on_2b', 'runner_on_3b',
    'score_differential', 'batter_hits_left', 'pitcher_throws_left', 'custom_zone_class'
]

batch_size = 32
train_dataset = PitchSequenceDataset(seq_train, pt_train, zone_train, strike_train, input_feature_cols)
dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, collate_fn=collate_fn)
```

```python
# Define model
import torch
import torch.nn as nn
from torch.nn.utils.rnn import pack_padded_sequence, pad_packed_sequence

class PitchLSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, 
                 num_pitch_types, num_zones, num_pitchers, embedding_dim=4):
        super(PitchLSTM, self).__init__()

        # Embed pitcher ID
        self.pitcher_embedding = nn.Embedding(num_pitchers, embedding_dim)

        # Updated input dim for LSTM
        self.lstm_input_dim = input_dim + embedding_dim

        # LSTM receives concatenated inputs
        self.lstm = nn.LSTM(self.lstm_input_dim, hidden_dim, num_layers, batch_first=True)

        # Multi-task outputs
        self.fc_pitch_type = nn.Linear(hidden_dim, num_pitch_types)
        self.fc_zone = nn.Linear(hidden_dim, num_zones)
        self.fc_is_strike = nn.Linear(hidden_dim, 1)

    def forward(self, x, lengths, pitcher_ids):
        # x: (batch_size, seq_len, input_dim)
        # pitcher_ids: (batch_size,)

        # Embed pitcher ID and repeat across sequence
        pitcher_embeds = self.pitcher_embedding(pitcher_ids)  # (batch_size, embedding_dim)
        pitcher_embeds = pitcher_embeds.unsqueeze(1).expand(-1, x.size(1), -1)  # (batch_size, seq_len, embedding_dim)

        # Concatenate pitcher embedding to each pitch in sequence
        x = torch.cat((x, pitcher_embeds), dim=2)  # (batch_size, seq_len, input_dim + embed_dim)

        # Pack and run through LSTM
        packed_input = pack_padded_sequence(x, lengths.cpu(), batch_first=True, enforce_sorted=True)
        packed_output, (hn, cn) = self.lstm(packed_input)

        last_hidden = hn[-1]  # Last layer's hidden state

        pitch_type_logits = self.fc_pitch_type(last_hidden)
        zone_logits = self.fc_zone(last_hidden)
        is_strike_logit = self.fc_is_strike(last_hidden)

        return pitch_type_logits, zone_logits, is_strike_logit
```

```python
input_dim = len(input_feature_cols)  
hidden_dim = 64
num_layers = 1
embedding_dim = 4  # for pitcher_id embedding
num_pitch_types = len(pitch_type_to_idx)
num_zones = len(set(y_zone))
num_pitchers = len(pitcher_to_idx)

# Initialize Model.
model = PitchLSTM(input_dim, hidden_dim, num_layers, 
                  num_pitch_types, num_zones,
                  num_pitchers, embedding_dim)
```

```python
# Define losses
criterion_pitch_type = nn.CrossEntropyLoss()
criterion_zone = nn.CrossEntropyLoss()
criterion_is_strike = nn.BCEWithLogitsLoss()

optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
# ---- Wrap Validation Set in Dataset and Dataloader ----
val_dataset = PitchSequenceDataset(seq_val, pt_val, zone_val, strike_val, input_feature_cols)

val_dataloader = DataLoader(
    val_dataset,
    batch_size=32, 
    shuffle=False,
    collate_fn=collate_fn
)
```

```python
num_epochs = 200

losses = []
zonelosses = []
strikelosses = []
typelosses = []
val_losses = []

best_val_loss = float('inf')
best_model_path = 'Best_Model.pt'

for epoch in range(num_epochs):
    total_loss = 0.0
    total_pitch_loss = 0.0
    total_zone_loss = 0.0
    total_strike_loss = 0.0

    for batch in dataloader:
        input_seq, lengths, pitch_target, zone_target, strike_target, pitcher_ids = batch

        pitch_target = pitch_target.long()
        zone_target = zone_target.long()
        strike_target = strike_target.float()

        # Forward
        pitch_logits, zone_logits, strike_logit = model(input_seq, lengths, pitcher_ids)

        # Compute Losses
        loss_pitch = criterion_pitch_type(pitch_logits, pitch_target)
        loss_zone = criterion_zone(zone_logits, zone_target)
        loss_strike = criterion_is_strike(strike_logit.view(-1), strike_target.view(-1))

        loss = loss_pitch + loss_zone + loss_strike
        total_loss += loss.item()
        total_pitch_loss += loss_pitch.item()
        total_zone_loss += loss_zone.item()
        total_strike_loss += loss_strike.item()
        
        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    avg_loss = total_loss / len(dataloader)
    avg_pitch_loss = total_pitch_loss / len(dataloader)
    avg_zone_loss = total_zone_loss / len(dataloader)
    avg_strike_loss = total_strike_loss / len(dataloader)
    
    losses.append(avg_loss)
    typelosses.append(avg_pitch_loss)
    zonelosses.append(avg_zone_loss)
    strikelosses.append(avg_strike_loss)

    print(f"Epoch {epoch+1}/{num_epochs} | Loss: {avg_loss:.4f}")
    print(f"  PitchTypeLoss: {avg_pitch_loss:.4f} | ZoneLoss: {avg_zone_loss:.4f} | StrikeLoss: {avg_strike_loss:.4f}")

    avg_len = lengths.float().mean().item()
    print(f"Batch avg length: {avg_len:.2f} | Loss: {loss.item():.4f}")

    # EVALUATE ON VALIDATION DATA
    model.eval()
    total_val_loss = 0.0

    with torch.no_grad():
        for batch in val_dataloader:
            input_seq, lengths, pitch_target, zone_target, strike_target, pitcher_ids = batch

            pitch_target = pitch_target.long()
            zone_target = zone_target.long()
            strike_target = strike_target.float()

            pitch_logits, zone_logits, strike_logit = model(input_seq, lengths, pitcher_ids)

            loss_pitch = criterion_pitch_type(pitch_logits, pitch_target)
            loss_zone = criterion_zone(zone_logits, zone_target)
            loss_strike = criterion_is_strike(strike_logit.view(-1), strike_target.view(-1))
            loss = loss_pitch + loss_zone + loss_strike

            total_val_loss += loss.item()

    avg_val_loss = total_val_loss / len(val_dataloader)
    val_losses.append(avg_val_loss)

    print(f"Epoch {epoch+1}/{num_epochs} | Train Loss: {avg_loss:.4f} | Val Loss: {avg_val_loss:.4f}")

    # Save best model if val loss improves
    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        torch.save(model.state_dict(), best_model_path)
        print(f"Saved new best model at epoch {epoch+1} with val loss {avg_val_loss:.4f}")
```

```python
# Evaluation with sigmoids on validation data:
from sklearn.metrics import f1_score

# ---- Load Model Weights ----
model.load_state_dict(torch.load('Best_Model.pt'))
model.eval()

# ---- Evaluate ----
correct_pitch, correct_zone, correct_strike = 0, 0, 0
total = 0

pitch_preds = []
pitch_trues = []

zone_preds = []
zone_trues = []

strike_preds = []
strike_trues = []
strike_probs = []  # Raw sigmoid outputs

threshold = 0.525 # Set your threshold here

with torch.no_grad():
    for batch in val_dataloader:
        input_seq, lengths, pitch_target, zone_target, strike_target, pitcher_id = batch

        pitch_target = pitch_target.long()
        zone_target = zone_target.long()
        strike_target = strike_target.float()

        pitch_logits, zone_logits, strike_logit = model(input_seq, lengths, pitcher_id)

        # ---- Predictions
        pitch_pred = pitch_logits.argmax(dim=1)
        zone_pred = zone_logits.argmax(dim=1)

        probs = torch.sigmoid(strike_logit).squeeze()  # Probabilities
        strike_pred = (probs > threshold).int()

        # ---- Append results
        pitch_preds.extend(pitch_pred.cpu().numpy())
        pitch_trues.extend(pitch_target.cpu().numpy())

        zone_preds.extend(zone_pred.cpu().numpy())
        zone_trues.extend(zone_target.cpu().numpy())

        strike_preds.extend(strike_pred.cpu().numpy())
        strike_trues.extend(strike_target.int().cpu().numpy())
        strike_probs.extend(probs.cpu().numpy())  # Keep raw probs for plots

# ---- Metrics
correct_pitch = sum(p == t for p, t in zip(pitch_preds, pitch_trues))
correct_zone = sum(p == t for p, t in zip(zone_preds, zone_trues))
correct_strike = sum(p == t for p, t in zip(strike_preds, strike_trues))
total = len(pitch_trues)

f1 = f1_score(strike_trues, strike_preds)
print(f"Strike F1 Score: {f1:.2f}")

# ---- Accuracy
print(f"Pitch Type Accuracy: {correct_pitch / total:.2%}")
print(f"Zone Accuracy: {correct_zone / total:.2%}")
print(f"Strike Accuracy: {correct_strike / total:.2%}")
```
